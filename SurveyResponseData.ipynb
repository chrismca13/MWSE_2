```python
import pandas as pd
import numpy as np
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
leaders_Deputy = pd.read_csv('Leadership Team - Deputy Director Performance Managment.csv')\
                   .drop(axis = 0, index = 0) # Drop first row with junk / meta data

leaders_Deputy['Respondent_Group'] = 'Leaders Toward Deputy Director'
```

```python
leaders_Director = pd.read_csv('Leadership Team - Executive Director Performance Managment.csv')\
                   .drop(axis = 0, index = 0) 

leaders_Director['Respondent_Group'] = 'Leaders Toward Executive Director'
```

```python
staff_Director = pd.read_csv('MWSE - Executive Director Performance Managment.csv')\
                   .drop(axis = 0, index = 0) 

staff_Director['Respondent_Group'] = 'Staff Toward Executive Director'
```

```python
staff_Deputy = pd.read_csv('MWSE - Deputy Director Performance Managment.csv')\
                   .drop(axis = 0, index = 0) 

staff_Deputy['Respondent_Group'] = 'Staff Toward Deputy Director'
```

```python
files = [leaders_Deputy, leaders_Director, staff_Director, staff_Deputy]
```

```python
df_all = pd.concat(files)
df_all.head()
```

```python
df_all_melt = df_all.melt(id_vars = ['Respondent ID', 'Collector ID', 'Start Date', 'End Date', 'IP Address',
                                     'Email Address', 'First Name', 'Last Name', 'Custom Data 1', 'Respondent_Group']   
                        , var_name = 'Question'
                        , value_name = 'Response')
```

```python
df_all_melt
```

```python
df_all_melt['Response'] = df_all_melt['Response'].fillna(0)
```

```python
df_all_text = df_all_melt[df_all_melt['Question'] == 'Do you have any additional feedback?']
df_all_response = df_all_melt[df_all_melt['Question'] != 'Do you have any additional feedback?']
```

```python
df_all_response['Negative_Feedback'] = np.where(df_all_response['Response'].astype(int) <= 2, 1,0)
df_all_response['Positive_Feedback'] = np.where(df_all_response['Response'].astype(int) >= 4, 1,0)
```

```python
df_all_response.to_csv('Numerical_Response.csv', index = False)
df_all_text.to_csv('Text_Response.csv', index = False)
```

```python
df_all_text = df_all_text[df_all_text['Response'] != 0]
```

```python
df_all_text['Respondent_Group'].unique()
```

```python
# Massage text responses to make word cloud

words = []

for response in df_all_text['Response']:
    a = response.split(" ")
    words.append(a)
    

flat_list = [item for sublist in words for item in sublist]
```

```python
word_cloud = pd.DataFrame(flat_list, columns = ['Words'])
word_cloud.to_csv('word_cloud.csv', index = False)
```
